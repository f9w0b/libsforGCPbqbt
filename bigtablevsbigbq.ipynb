```python
# import google cloud bigtable libraries 
from google.cloud import bigtable
from google.cloud.bigtable import column_family
from google.cloud.bigtable import row_filters
import pandas as pd
```

```python
# Use the function:
bt_proj_id = 'grounded-gizmo-394701'
bt_inst_id = 'apache-poc-bt1994'
table_name = 'test-bt-model12'
```

```python
def printBigtableRows(row):
    print("Row key: {}".format(row.row_key.decode('utf-8')))
    
    for column_fam, cols in sorted(row.cells.items()):
        print("Column Family {}".format(column_fam))
        for col, cells in sorted(cols.items()):
            for cell in cells:
                col_name = col.decode('utf-8')
                timestamp = cell.timestamp
                value = cell.value.decode('utf-8')
                print("\t{}:  {}   ||| Time : {}".format(col_name, value, timestamp))
```

```python
def readBigtable(bt_proj_id, bt_inst_id, table_name):
    client = bigtable.Client(project = bt_proj_id, admin = True)
    instance = client.instance(bt_inst_id)
    #connect to table 
    table = instance.table(table_name)
    
    # Create a filter to limit the data returned.
    # For instance, this filter fetches only the latest version of each cell.
    filter_ = row_filters.CellsColumnLimitFilter(1)
    
    rows = table.read_rows(filter_=filter_)
    # print(f"Total rows fetched: {len(list(rows))}")
    # use the function above
    for r in rows:
        printBigtableRows(r)
    # print(f"Total rows fetched: {len(list(rows))}")
```

```python
readBigtable(bt_proj_id, bt_inst_id, table_name)
```

```python
def bigtabletoDataframe(bt_proj_id, bt_inst_id, table_name, filter_number):
    # Connect to Bigtable
    client = bigtable.Client(project=bt_proj_id, admin=True)
    instance = client.instance(bt_inst_id)
    table = instance.table(table_name)
    
    filter_ = row_filters.CellsColumnLimitFilter(filter_number)
    # Read rows
    rows = table.read_rows(filter_= filter_)
    # print(f"Total rows fetched: {len(list(rows))}")
    # rows.consume_all()
    # print(f"Total rows fetched: {len(list(rows))}")
    
    # Convert rows to a list of dictionaries
    data = []
  
    for row in rows:
            row_data = {"row_key": row.row_key.decode('utf-8')}
            print("Row key: {}".format(row.row_key.decode('utf-8')))
            for column_fam, cols in sorted(row.cells.items()):
                print("Column Family {}".format(column_fam))
                for col, cells in sorted(cols.items()):
                    for cell in cells:
                        col_name = col.decode('utf-8')
                        timestamp = cell.timestamp
                        value = cell.value.decode('utf-8')
                        print("\t{}:  {}   ||| Time : {}".format(col_name, value, timestamp))
                        row_data.update({f"{col_name}" : value,"Timestamp": timestamp}) 
            
            data.append(row_data)
        
                
            # row_data = {"row_key": row.row_key.decode('utf-8')}
            # print(row_data)
            # for column_family, columns in row.cells.items():
            #     for column, cells in columns.items():
            #         column_name = column.decode('utf-8').split(':')[-1]  # Assumes column name format is cf:column_name
            #         row_data[column_name] = cells[0].value.decode('utf-8')  # Fetches the latest cell value
            # data.append(row_data)
        
    
    # print(f"Total rows fetched: {len(list(rows))}")
    
    # Convert list of dictionaries to DataFrame
    df = pd.DataFrame(data)
    
    return df


             
```
