```python
def get_undelivered_message_count():
    # Initialize a Pub/Sub subscription client
    subscriber = pubsub_v1.SubscriberClient()
    
    # Construct the subscription path
    subscription_path = subscriber.subscription_path(project_id, subscription_id)
    
    # Retrieve subscription details
    subscription = subscriber.get_subscription(request={"subscription": subscription_path})
    
    # Extract num_undelivered_messages attribute
    undelivered_count = subscription.num_undelivered_messages
    
    return undelivered_count
```
